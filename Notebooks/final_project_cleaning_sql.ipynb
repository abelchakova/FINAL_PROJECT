```python
# Drop irrelevant columns
columns_to_drop = ['Unnamed: 0', 'title', 'sq_mt_useful', 'latitude', 'longitude', 'is_exact_address_hidden', 
                   'street_name', 'street_number', 'raw_address', 'portal', 'operation', 'is_floor_under', 'door', 
                   'rent_price_by_area', 'is_rent_price_known', 'is_buy_price_known', 'buy_price_by_area', 'is_furnished', 'is_kitchen_equipped',
                   'is_new_development', 'are_pets_allowed', 'is_exterior', 'is_accessible', 'rent_price',
                   'has_central_heating', 'has_individual_heating', 'energy_certificate', 'has_private_parking','has_public_parking']
housing_madrid.drop(columns=columns_to_drop, inplace=True)
housing_madrid.isna().sum()
```

```python
# Handling missing values

# 1. Drop rows with missing values in columns 'sq_mt_built' and 'n_bathrooms'
housing_madrid.dropna(subset=['sq_mt_built', 'n_bathrooms'], inplace=True)

# 2. Replace missing values in columns 'n_floors', 'parking_price' and 'sq_mt_allotment' with 0
housing_madrid['n_floors'].fillna(0, inplace=True)
housing_madrid['parking_price'].fillna(0, inplace=True)
housing_madrid['sq_mt_allotment'].fillna(0, inplace=True)

# 3. Replace missing values in column 'house_type_id' with 'Unknown'
housing_madrid['house_type_id'].fillna('Unknown', inplace=True)

# 4. Replace missing values in column 'built_year' with average
average_built_year = housing_madrid['built_year'].mean()
housing_madrid['built_year'].fillna(average_built_year, inplace=True)

# 5. Replace missing values in column 'floor' with mode
floor_mode = housing_madrid['floor'].mode()[0]
housing_madrid['floor'].fillna(floor_mode, inplace=True)

# 6. Update a specific value in the 'built_year' column
# 6.1. Identify the outlier value
outlier_value = 8170
new_value = 1970

# 6.2. Replace the outlier value with the new value
housing_madrid.loc[housing_madrid['built_year'] == outlier_value, 'built_year'] = new_value

# Verify the changes
print(housing_madrid.info())
```

```python
# Define a mapping for 'house_type_id' column
house_type_mapping = {
    'HouseType 1: Pisos': 'Apartment',
    'HouseType 4: Dúplex': 'Duplex',
    'HouseType 5: Áticos': 'Penthouse',
    'HouseType 2: Casa o chalet': 'House or Chalet'
}

# Apply the mapping to 'house_type_id' column to create a new 'house_type' column
housing_madrid['house_type'] = housing_madrid['house_type_id'].map(house_type_mapping)

# Define a new mapping for the new 'house_type' column to create 'house_type_id'
house_id_mapping = {
    'Apartment': 1,
    'Duplex': 2,
    'Penthouse': 3,
    'House or Chalet': 4
}

# Apply the mapping to 'house_type' column to create a new 'house_type_id' column
housing_madrid['house_type_id'] = housing_madrid['house_type'].map(house_id_mapping)

# Print unique values to check the mapping
print(housing_madrid['house_type'].unique())
print(housing_madrid['house_type_id'].unique())
```

```python
# Convert Data Types to relevant

# Convert columns to integer
columns_to_integer = ['n_bathrooms', 'n_floors', 'built_year', 'floor']
housing_madrid[columns_to_integer] = housing_madrid[columns_to_integer].map(int)

# Convert columns to float
columns_to_float = ['buy_price']
housing_madrid[columns_to_float] = housing_madrid[columns_to_float].astype(float)

# Convert columns to boolean
columns_to_boolean = [
    'has_ac', 'has_fitted_wardrobes', 
    'has_lift', 'has_garden', 'has_pool', 'has_terrace', 'has_balcony', 'has_storage_room', 
    'has_green_zones', 'is_parking_included_in_price', 
    'is_orientation_north', 'is_orientation_west', 'is_orientation_south', 'is_orientation_east'
]

# Convert 'True'/'False' strings to boolean values
housing_madrid[columns_to_boolean] = housing_madrid[columns_to_boolean].map(lambda x: True if x == 'True' or x == True else False)

# Check data types after conversion
housing_madrid.dtypes
```

```python
# Creating 3 columns out of neighborhood_id
import re

# Define function to extract neighborhood ID
def extract_neighborhood_id(value):
    match = re.search(r'Neighborhood (\d+):', str(value))
    return int(match.group(1)) if match else None

# Define function to extract district ID
def extract_district_id(value):
    match = re.search(r'District (\d+):', str(value))
    return int(match.group(1)) if match else None

# Define function to extract district name
def extract_district_name(value):
    match = re.search(r'District \d+: (.+)$', str(value))
    return match.group(1) if match else None

# Apply functions to create new columns
housing_madrid['district_id'] = housing_madrid['neighborhood_id'].apply(extract_district_id)
housing_madrid['district'] = housing_madrid['neighborhood_id'].apply(extract_district_name)
housing_madrid['neighborhood_id'] = housing_madrid['neighborhood_id'].apply(extract_neighborhood_id)

# Display the updated DataFrame
housing_madrid.head()
```
